# Introducing Linear Regression

- Linear regression is a foundational method in machine learning and statistics.
- It models the relationship between:
  - A continuous response variable (output)
  - One or more explanatory variables (inputs)
- The model assumes a **linear relationship** between inputs and output.
- Parameters are typically estimated using **least squares optimization**.

---

## Simple Linear Regression

- Models the relationship between:
  - One predictor variable
  - One response variable
- Assumes the expected value of the response changes linearly with the input.
- Useful for understanding basic trends and relationships.

**Model:**

$$
y = \beta_0 + \beta_1 x
$$

- $\beta_0$: intercept  
- $\beta_1$: slope (effect of $x$ on $y$)  
- $y$: predicted response

---

## Multiple Linear Regression

- Extends simple linear regression to multiple predictors.
- Models the combined effect of several input variables.
- Each feature contributes linearly to the prediction.
- Commonly used in real-world data analysis.

**Model:**

$$
y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_p x_p
$$

- $x_1, x_2, \dots, x_p$: input features  
- $\beta_1, \beta_2, \dots, \beta_p$: coefficients  
- $\beta_0$: intercept

## Key Takeaways

- Both models assume linearity between inputs and output.
- The goal is to estimate coefficients that minimize prediction error.
- Linear regression serves as the foundation for:
  - Regularized regression (Ridge, Lasso)
  - Generalized linear models
  - Many modern machine learning methods

# Linear Regression Using Stochastic Gradient Descent (SGD)

## Linear Regression via Minibatch SGD

**1. Setup**
- Define per-example squared loss: $l^{(i)}(\mathbf{w}, b) = \frac{1}{2} \left(\hat{y}^{(i)} - y^{(i)}\right)^2$ where $\hat{y}^{(i)} = \mathbf{w}^\top \mathbf{x}^{(i)} + b$
- Define empirical loss: $L(\mathbf{w}, b) = \frac{1}{n} \sum_{i=1}^n \frac{1}{2}\left(\mathbf{w}^\top \mathbf{x}^{(i)} + b - y^{(i)}\right)^2$
- Initialize $\mathbf{w}$ and $b$ randomly

**2. Training Loop** (repeat until convergence)
- Sample minibatch $\mathcal{B}_t$ of size $|\mathcal{B}|$
- Update parameters with learning rate $\eta$:

$$
\mathbf{w} \leftarrow \mathbf{w} - \frac{\eta}{|\mathcal{B}|} \sum_{i \in \mathcal{B}_t} \mathbf{x}^{(i)} \left(\mathbf{w}^\top \mathbf{x}^{(i)} + b - y^{(i)}\right)
$$

$$
b \leftarrow b - \frac{\eta}{|\mathcal{B}|} \sum_{i \in \mathcal{B}_t} \left(\mathbf{w}^\top \mathbf{x}^{(i)} + b - y^{(i)}\right)
$$

**3. Prediction**
- For new input $\mathbf{x}_{\text{new}}$: $\hat{y}_{\text{new}} = \hat{\mathbf{w}}^\top \mathbf{x}_{\text{new}} + \hat{b}$

---
#### Analytic Solution (Closed-Form)

**Derivation:**
- Absorb bias into $\mathbf{w}$ by appending column of 1s to $\mathbf{X}$
- Minimize $\|\mathbf{y} - \mathbf{X}\mathbf{w}\|^2$
- Take derivative and set to zero:

$$
\partial_{\mathbf{w}} \|\mathbf{y} - \mathbf{X}\mathbf{w}\|^2 = 2 \mathbf{X}^\top (\mathbf{X} \mathbf{w} - \mathbf{y}) = 0
$$

- Rearrange: $\mathbf{X}^\top \mathbf{y} = \mathbf{X}^\top \mathbf{X} \mathbf{w}$
- Solve for $\mathbf{w}$:

$$
\mathbf{w}^* = (\mathbf{X}^\top \mathbf{X})^{-1}\mathbf{X}^\top \mathbf{y}
$$

**Note:** Requires $\mathbf{X}^\top \mathbf{X}$ invertible (full rank). While this gives exact solutions for linear regression, most deep learning problems lack analytic solutions‚Äîmaking gradient descent essential.

## ‚öôÔ∏è Gradient Descent Linear Regression Estimator Implementation

````python
class LinearRegressionGD(BaseEstimator, RegressorMixin):  # ‚Üê ADD INHERITANCE HERE
    def __init__(self, eta=0.01, n_iter=50, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state

    def fit(self, X, y):
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=X.shape[1])
        self.b_ = np.array([0.])
        self.losses_ = []
        
        self.n_features_in_ = X.shape[1]  # ‚Üê ADD THIS ONE LINE (critical for sklearn validation)

        for i in range(self.n_iter):
            output = self.net_input(X)
            errors = (y - output)
            self.w_ += self.eta * 2.0 * X.T.dot(errors) / X.shape[0]
            self.b_ += self.eta * 2.0 * errors.mean()
            loss = (errors**2).mean()
            self.losses_.append(loss)
        return self

    def net_input(self, X):
        return np.dot(X, self.w_) + self.b_

    def predict(self, X):
        return self.net_input(X)
````

# Streamlining workflows with pipelines

## üßπ Feature Preprocessing Pipeline

````python
# Build NUMERIC pipeline: impute ‚Üí scale
numeric_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),  # median robust to outliers
    ('scaler', StandardScaler())
])

# Build CATEGORICAL pipeline: impute ‚Üí encode
categorical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))  # critical for production
])

# Combine in ColumnTransformer
preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_pipeline, numeric_features),
    ('cat', categorical_pipeline, categorical_features)
])
````

## üéØ Final Model Evaluation: Train vs. Test Performance Metrics

### üìä Mean Squared Error (MSE)

- Mean Squared Error (MSE) is a quantitative measure of a model‚Äôs performance.

- It is the loss function minimized when fitting a linear regression model.

- MSE measures the average squared difference between observed values and predicted values:

  $$
  \text{MSE} = \frac{1}{n}\sum_{i=1}^{n}\left(y^{(i)} - \hat{y}^{(i)}\right)^2
  $$

- This formulation does not include the \( \frac{1}{2} \) scaling factor often used in gradient descent.

- A smaller MSE indicates better predictive accuracy.



````python
# Generate predictions on training and test sets using the optimized model
y_pred_train = reg_sgd.predict(X_train)
y_pred_test = reg_sgd.predict(X_test)

# Compute metrics
rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
r2_train = r2_score(y_train, y_pred_train)
r2_test = r2_score(y_test, y_pred_test)

# Print results (clean formatting - YOUR preferred style)
print(f"\n=== FINAL MODEL PERFORMANCE ===")
print(f"RMSE train: {rmse_train:.4f}")
print(f"RMSE test : {rmse_test:.4f}")
rmse_gap_pct = (rmse_test - rmse_train) / rmse_train * 100 if rmse_train > 0 else 0
print(f"Gap       : {rmse_gap_pct:+.1f}% {'‚Üë OVERFITTING' if rmse_gap_pct > 15 else ''}")

print(f"\nR¬≤   train: {r2_train:.4f}")
print(f"R¬≤   test : {r2_test:.4f}")
r2_gap = r2_train - r2_test
print(f"Gap       : {r2_gap:+.4f} {'‚Üì OVERFITTING' if r2_gap > 0.1 else ''}")

# Diagnostic for negative R¬≤ (pedagogical teaching moment!)
if r2_test < 0:
    print(f"\n‚ö†Ô∏è  NEGATIVE R¬≤ DETECTED")
    print(f"   Model performs worse than predicting constant = {y_train.mean():.2f}")
    print(f"   ‚Üí Check: feature scaling? learning rate too high? iterations too low?")
````

## üìê Visualizing Feature-Target Relationship: Partial Dependence Plot

````python
def lin_regplot_scaled(X_train, y_train, pipeline, feature_name, numeric_features):
    # Extract pipeline components
    preprocessor = pipeline.named_steps['preprocessor']
    regressor = pipeline.named_steps['reg_SGD']
    
    # Transform X and scale y
    X_transformed = preprocessor.transform(X_train)
    feature_idx = list(numeric_features).index(feature_name)
    X_feature = X_transformed[:, feature_idx]
    
    sc_y = StandardScaler()
    y_scaled = sc_y.fit_transform(y_train.values.reshape(-1, 1)).flatten()
    
    # Create predictions (vary feature, hold others at mean)
    X_plot = np.tile(X_transformed.mean(axis=0), (len(X_feature), 1))
    X_plot[:, feature_idx] = X_feature
    
    # Sort and predict
    sort_idx = X_feature.argsort()
    y_pred = sc_y.transform(regressor.predict(X_plot[sort_idx]).reshape(-1, 1)).flatten()
    
    # Plot
    plt.scatter(X_feature, y_scaled, c='steelblue', edgecolor='white', s=70)
    plt.plot(X_feature[sort_idx], y_pred, color='black', lw=2)
    plt.xlabel(f'{feature_name} (standardized)')
    plt.ylabel('Sale price (standardized)')
    return plt
````

## üîß Implementing a Custom SVD-Based Linear Regression Estimator

````python
class LinearRegressionSVD(BaseEstimator, RegressorMixin):  # ‚Üê ADD INHERITANCE
    def __init__(self):
        pass  # No parameters to store (minimal change)
    
    def fit(self, X, y):
        # ADD THIS ONE LINE for sklearn validation (required by check_estimator)
        self.n_features_in_ = X.shape[1]
        
        # Add bias column to X (column of ones)
        X_with_bias = np.c_[np.ones(X.shape[0]), X]
        
        # Solve using SVD: X_with_bias @ w = y
        U, S, VT = np.linalg.svd(X_with_bias, full_matrices=False)
        
        # Compute weights: w = VT.T @ inv(S) @ U.T @ y
        w_with_bias = VT.T @ np.linalg.inv(np.diag(S)) @ U.T @ y
        
        # Separate bias and weights
        self.b_ = np.array([w_with_bias[0]])
        self.w_ = w_with_bias[1:]
        
        # ADD THIS ONE LINE for sklearn compatibility (required by RegressorMixin.score())
        self.n_features_in_ = X.shape[1]  # Already set above, but shown for clarity
        
        return self
    
    def net_input(self, X):
        return np.dot(X, self.w_) + self.b_
    
    def predict(self, X):
        return self.net_input(X)
````
